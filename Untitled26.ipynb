```python
import numpy as np
import requests
from scipy.stats import ttest_ind, mannwhitneyu

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
import matplotlib.pyplot as plt

Loadeddata = pd.read_csv("traffic_index.csv")

print("Head Of Data= ")
print(Loadeddata.head())
Loadeddata['trafficindexdate'] = pd.to_datetime(Loadeddata['trafficindexdate'], errors='coerce')
Loadeddata = Loadeddata.dropna(subset=['trafficindexdate'])


print("Information about data= ")
Loadeddata.info()

plt.figure(figsize=(12,6))

sns.histplot(Loadeddata['average_traffic_index'], bins=20, kde=True)

plt.title("Average Traffic Indexing Dist")

plt.xlabel("Average Traffic Indexing")


plt.ylabel("All Days")

plt.show()
```

So, in this project I generally used a traffic collection with 3332 amount daily savings from Istanbul city. A sample of the data illustrates that the daily average traffic index values are from 23 to 58, that illustrates that the amount of congestion changes and replaces itself within times. Based on that trend, it looks like traffic amounts are regular, but really bad ones do not happen very often.

```python
temp = Loadeddata['trafficindexdate'].min()
ST = temp.strftime('%Y-%m-%d')
temp2 = Loadeddata['trafficindexdate'].max()
ED = temp2.strftime('%Y-%m-%d')

y = 28.9784
cnt = 0
x = 41.0082

#Api part
Fetch = (
    f"https://archive-api.open-meteo.com/v1/archive?"
    f"latitude={x}&longitude={y}"
    f"&start_date={ST}&end_date={ED}"
    f"&daily=temperature_2m_max,temperature_2m_min,precipitation_sum,"
    f"windspeed_10m_max,relative_humidity_2m_mean"
    f"&timezone=Europe%2FIstanbul"
)

# Request from website now
got = requests.get(Fetch)
freshdata = got.json()
Wdataframe = pd.DataFrame(freshdata['daily'])
cnt = cnt + 1
Wdataframe['time'] = pd.to_datetime(Wdataframe['time'])
print("Head of weather data= ")
print(Wdataframe.head())
flag = True

Wdataframe.to_csv("weather.csv", index=False)
```

The Open-Meteo API was used to collect data about the weather in Istanbul with given coordinates. That API gives away free historical weather data freely. So I got daily averages for highest and lowest temperatures, total precipitation, wind speed, and relative humidity in Istanbul with a specific date range that fits with my traffic data's time range. That instance illustrates standard weather conditions for early August 2015, with highs around from 28 degree to 30 degrees, little to no rain, and mild wind. That API-based data collection method ensured that there was a steady and reliable source for appending to the traffic knowledge and checking how the weather can influence the ratio of traffic in Istanbul.

```python
#Now I will merge these two datas
Loadeddata['trafficindexdate'] = Loadeddata['trafficindexdate'].dt.tz_localize(None)

if(flag == True):
  Unified = pd.merge(Loadeddata, Wdataframe, left_on='trafficindexdate', right_on='time')
  Unified = Unified.drop(columns=['time'])
  Unified.to_csv("UnifiedData.csv", index=False)

print("Unified Data=")
print(Unified.head())
```

```python
#Hypothesis testing part


# Hypothesis 1 = Does rain increase traffic?
rainratio = 0.2
Unified['is_rainy'] = Unified['precipitation_sum'] > rainratio
hyptest = 0.05
weekends = 5
Rainexist = Unified[Unified['is_rainy']]['average_traffic_index']
Rainnotexist = Unified[~Unified['is_rainy']]['average_traffic_index']

#Now I do t-test
TVAL, PVALUE = ttest_ind(Rainexist, Rainnotexist, equal_var=False)

print("Hypothesis 1: Does rain increase traffic intensity?")
print(f"t-statistic= {TVAL:.4f}")
print(f"p-value= {PVALUE:.4f}")
if(PVALUE < hyptest):
    print("Statistically significant, rain likely increases traffic")
else:
    print("Not statistically significant, rain might not affect traffic")

#Second hypothesis is = Is weekend traffic different?
Unified['day_of_week'] = Unified['trafficindexdate'].dt.dayofweek
Unified['is_weekend'] = Unified['day_of_week'] >= weekends
notworkingday = Unified[Unified['is_weekend']]['average_traffic_index']
workingday = Unified[~Unified['is_weekend']]['average_traffic_index']
#Now doing Mann-Whitney U test
UVAL, PVALUEVERSION2 = mannwhitneyu(notworkingday, workingday, alternative='two-sided')

print("\nHypothesis 2: Is weekend traffic different from weekdays?")
print(f"U-statistic= {UVAL:.4f}")
print(f"p-value= {PVALUEVERSION2:.4f}")
if(PVALUEVERSION2 < hyptest):
    print("Statistically significant, weekend traffic differs from weekdays.")
else:
    print("Not statistically significant, no clear difference found.")
```

Two different statistical hypothesis tests I have done to simulate how weather and time of day influenced the traffic density. The first test checked whether more people drive on rainy days. The consequence is statistically significant, with a roughly t-statistic of 7.3605 and a p-value of 0. That means that rain is positively bound to high levels of average traffic index. The second hypothesis checked whether traffic trends are different on weekends when we compare them with weekdays. The Mann-Whitney U test illustrated a U-statistic of 467,841.5 and a p-value of 0; that generally means there is a significant difference in the traffic density on weekends when we compare it with weekdays. Those results support the idea that both the weather and the time of day influence the density of the roads.

```python
#ML Part
Lenof = 0.2
Randnum = 42
#target is that is this a high-traffic day?
Mcalculated = Unified['average_traffic_index'].median()
Unified['high_traffic'] = (Unified['average_traffic_index'] > Mcalculated).astype(int)
Scorecounter = 4
#Features definition
FF = ['temperature_2m_max', 'temperature_2m_min', 'precipitation_sum','windspeed_10m_max', 'relative_humidity_2m_mean']
xval = Unified[FF]
yval = Unified['high_traffic']
testfailedornot = False
# Train / test separation
Trainn, testx1, Trainny, testy1 = train_test_split(xval, yval, test_size=Lenof, random_state= Randnum )
# Training the random forest modelling
model = RandomForestClassifier(random_state=42)
model.fit(Trainn, Trainny)
#Now predict
Predicitionmethod = model.predict(testx1)
Probability = model.predict_proba(testx1)[:, 1]

# Evaluate model
print("Classification report=")
print(classification_report(testy1, Predicitionmethod))

print("Confusion matrix= ")
print(confusion_matrix(testy1, Predicitionmethod))

print("ROC AUC Score= ", round(roc_auc_score(testy1, Probability), Scorecounter))

testfailedornot = True
cruciality = pd.Series(model.feature_importances_, index=FF)
cruciality = cruciality.sort_values()
Scorecounter = 0
Randnum = 0
plt.figure(figsize=(8,5))
Lenof = 0
cruciality.plot(kind='barh')

plt.title("Feature importances in predicting high traffic")

plt.xlabel("Importance score")
plt.show()
```

A Random Forest Classifier was used as a main method to form an ML model that represents, from the weather, whether a specific day would have too much traffic. The model was 57% correct, and both its precision and recall scores were roughly around 0.57. That means it did a good job but not a perfect one. With an ROC AUC score of 0.5904, the model looks like it does a little better than selecting at random. The confusion matrix illustrates that the number of correct and incorrect guesses is about the same for the group, which is days with a lot and a few. The most crucial indicators, according to feature importance analysis, were temperature variables in both max and min and speed of wind. Humidity and precipitation came in second and third. That illustrates that temperature and wind parameters are more crucial than rain as alone in specifying how crowded the roads are.
